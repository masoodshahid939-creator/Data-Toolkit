Q1: What is the difference between multithreading and multiprocessing?
Answer: Multithreading means running different threads inside the same process. They share the same memory space and are useful when tasks involve waiting (like reading a file or network requests).
Multiprocessing means using different processes, and each process has its own memory. This allows true parallel execution, so it is better for heavy calculations.

```python
#Q3: Python program that logs an error message when division by zero happens.
import logging

logging.basicConfig(filename="error.log", level=logging.ERROR)

try:
    x = 10 / 0
except ZeroDivisionError as e:
    logging.error("Error occurred: %s", e)
```

Q6: What are the differences between NumPy arrays and Python lists?
Answer:NumPy arrays can only hold one type of data, while lists can hold mixed types.
NumPy is faster and uses less memory for numbers.
Mathematical operations can be done directly on NumPy arrays, but not on lists.
Lists are more flexible for general use, while NumPy is best for large data and math.

Q7: Difference between apply() and map() in Pandas.
Answer:map() works only on a single column (Series). It applies a function to each value.
apply() works on both Series and DataFrames. It can apply a function to a row or a column.
So, map() is simple, while apply() is more powerful and flexible.

```python
#Q10: Calculate correlation matrix and show it with heatmap.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "Math": [85, 78, 90, 72, 88],
    "Science": [80, 75, 85, 70, 82],
    "English": [78, 82, 88, 65, 80]
}

df = pd.DataFrame(data)

corr = df.corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```
